```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("poly_practice_2.csv")
```

```python
data.head()
```

```python
x=data.x
```

```python
y=data.y
```

```python
w0=.4
w1=.5
k=0
lr=.0001
while(k<100):
    yp=w0+w1*x
    loss=1/len(x)*sum((yp-y)**2)
    print(loss)
    w0=w0-lr*(2/len(x)*sum((yp-y)))
    w1=w1-lr*(2/len(x)*sum((yp-y)*x))
    plt.plot(x,y)
    plt.plot(x,yp)
    k=k+1
```

```python
w0=.4
w1=.5
w2=.7
k=0
lr=.00000001
while(k<100):
    yp=w0+w1*x+w2*x**2
    loss=1/len(x)*sum((yp-y)**2)
    print(loss)
    w0=w0-lr*(2/len(x)*sum((yp-y)))
    w1=w1-lr*(2/len(x)*sum((yp-y)*x))
    w2=w1-lr*(2/len(x)*sum((yp-y)*x**2))
    plt.plot(x,y)
    plt.plot(x,yp)
    k=k+1
```

```python
w0=.4
w1=.5
w2=.7
w3=.9
k=0
lr=.000000000001
while(k<100):
    yp=w0+w1*x+w2*x**2+w3*x**3
    loss=1/len(x)*sum((yp-y)**2)
    print(loss)
    w0=w0-lr*(2/len(x)*sum((yp-y)))
    w1=w1-lr*(2/len(x)*sum((yp-y)*x))
    w2=w1-lr*(2/len(x)*sum((yp-y)*x**2))
    w3=w1-lr*(2/len(x)*sum((yp-y)*x**3))
    plt.plot(x,y)
    plt.plot(x,yp)
    k=k+1
```
