```python
def outliers(data):
    import numpy as np 
    for i in range(len(data.columns)):
        y=data.iloc[:,i]
        out=[]
        mean=np.mean(y)
        std=np.std(y)
        for j in y:
            z=(j-mean)/std
            if(z>3):
                out.append(j)
        for m in out:
            data = data[data.iloc[:,i]!=m]
    return data
data3 = outliers(data2)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
def calc_vif(X):

    # Calculating VIF
    vif = pd.DataFrame()
    vif["columns"] = X.columns
    vif["VIF_Score"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

    return(vif)

m=calc_vif(x1)
```
